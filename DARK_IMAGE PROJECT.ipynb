# Project Title:- Storing image in Matrix and converting image in dark mode

```python
import numpy as np
import matplotlib.pyplot as plt
```

let's suppose we have a image and we want to modify it in the numpy so, the pixels, dimensions and other numerical properties of that image will save as in the large numpy array form and we have to save it and load it in the future.

we will also save our array in the memory by using np.save

the extension is .npy

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.random.rand(3, 3)
array3 = np.zeros((4, 4))


np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

* load the array-> np.load

```python
loaded_array1 = np.load('array1.npy')
loaded_array2 = np.load('array2.npy')
loaded_array3 = np.load('array3.npy')
print(loaded_array1)
print(loaded_array2)
print(loaded_array3)
```

### The same thing I will do with the numpy logo
* first save the image as .npy format
* then load the data
* then display the data
* finally make the dark version of the image

```python
# load the data using try and except
try:
    loaded_image = np.load('numpy_logo.npy')

    # display the data
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(loaded_image)
    plt.title("NumPy Logo")
    plt.grid(False)

    dark_logo = 1 - loaded_image
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("NumPy Dark Logo")
    plt.grid(False)

except FileNotFoundError:
    print("Numpy logo File not found")
```

```python
try:
    loaded_image = np.load('numpy_logo.npy')

    # Ensure image is displayed with a white background
    if loaded_image.shape[-1] == 4:  # If it has an alpha channel
        loaded_image = loaded_image[:, :, :3]  # Remove alpha (RGBA to RGB)

    plt.figure(figsize=(10, 5))

    plt.subplot(121)
    plt.imshow(loaded_image)
    plt.title("NumPy Logo")
    plt.grid(False)  # Remove grid/axes

    dark_logo = 1 - loaded_image
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("NumPy Dark Logo")
    plt.grid(False)  # Remove grid/axes

    plt.show()

except FileNotFoundError:
    print("Numpy logo file not found")
```
